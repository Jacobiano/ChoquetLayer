```python
BATCH_SIZE = int(32.)
EPOCHS = int(512.)

batch_size = BATCH_SIZE
epochs = EPOCHS
learning_rate=0.001
CHANNELS=1
NLAYERS=2
SHRINK=1
NFILTERS=48
KSIZE=13
SUBSPACE=12
PATIENCE_ES=40
PATIENCE_RP=5


class NeymanScott:
    """
    Neyman-Scott point process using a Poisson variable for the number of parent points, uniform for
    the number of daughter points and Pareto distribution for the distance from the daughter points to
    the parent.
    """
    def __init__(self,
                 poisson_mean: float,
                 daughter_max: int,
                 pareto_alpha: float,
                 pareto_scale: float,
                 size: (int, int)):
        """
        :param poisson_mean: mean of the number of parent points
        :param daughter_max: maximum number of daughters per parent points
        :param pareto_alpha: alpha parameter of the Pareto distribution
        :param pareto_scale: scale used in the Pareto distribution. This parameter is
            applied before resizing the points from the [0, 1] interval to the size of the image.
        :param size: rescale the output to this size
        """
        self.poisson_mean = poisson_mean
        self.daughter_max = daughter_max
        self.pareto_alpha = pareto_alpha
        self.pareto_scale = pareto_scale
        self.size = np.array([size])
        self.generator = np.random.Generator(np.random.PCG64())

    def __call__(self):
        num_parents = self.generator.poisson(lam=self.poisson_mean)
        parents = self.generator.random((num_parents, 2))
        num_daughters = self.generator.integers(1, self.daughter_max, num_parents)
        points = np.empty((0, 2))

        for i in range(num_parents):
            # normalizes the pareto II distribution
            dist = self.generator.pareto(self.pareto_alpha, (num_daughters[i], 1))
            dist = (dist + 1) * self.pareto_scale
            angle = self.generator.uniform(0., 2 * np.pi, (num_daughters[i],))
            positions = np.stack([np.cos(angle), np.sin(angle)], 1)
            positions *= dist
            positions += parents[i, np.newaxis, :]
            points = np.concatenate([points, positions])
        # remove points outside the set [0, 1] x [0, 1]
        valid_points = np.logical_and(
            np.logical_and(0. <= points[:, 0], points[:, 0] <= 1.),
            np.logical_and(0. <= points[:, 1], points[:, 1] <= 1.)
        )
        points = points[valid_points, :]
        # scale to the image size
        points = points * self.size
        return points



NSAMPLES_TRAINING=2024*2
IMG_SIZE=128

poisson_mean=100
daughter_max=50
pareto_scale=.02
pareto_alpha=1. #GENERATION ON IT
```

```python
def net(points,W1,W2=np.eye(2)):
    L1=np.matmul(points,W)
    L1=np.maximum(L1,0)
    L2=np.matmul(L1,W2)
    L2=np.maximum(L2,0)
    return L2
    
W=np.random.randn(2,2)
print(W)

nx, ny = (50, 50)
x = np.linspace(0, 128, 50)
y = np.linspace(0, 128, 50)
x, y = np.meshgrid(x, y)
x= x.ravel()
y= y.ravel()
Z=np.concatenate([np.expand_dims(x,axis=-1),np.expand_dims(y,axis=-1)],axis=-1)
Z=net(Z,W)
plt.plot(Z[:,0],Z[:,1],'.')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.draw()
plt.show()

for i in range(10):
    pareto_alpha1=.1
    gen = NeymanScott(poisson_mean, daughter_max, pareto_alpha1, pareto_scale, (IMG_SIZE, IMG_SIZE))
    points = gen()
    points = net(points,W)
    dist1 = cdist(points, Z, metric="euclidean" )
    
    
    pareto_alpha2=2.
    gen = NeymanScott(poisson_mean, daughter_max, pareto_alpha2, pareto_scale, (IMG_SIZE, IMG_SIZE))
    points = gen()
    points = np.matmul(points,W)
    dist2 = cdist(points, Z, metric="euclidean" )
    
    pareto_alpha3=6.
    gen = NeymanScott(poisson_mean, daughter_max, pareto_alpha3, pareto_scale, (IMG_SIZE, IMG_SIZE))
    points = gen()
    points = net(points,W)
    dist3 = cdist(points, Z, metric="euclidean" )

    v=np.max(np.min(dist1,axis=0))
    v=np.maximum(v,np.max(np.min(dist2,axis=0)))
    v=np.maximum(v,np.max(np.min(dist3,axis=0)))
    hist, bin_edges = np.histogram(np.concatenate([np.zeros(1),np.min(dist1,axis=0),np.ones(1)*v]), density=True)
    pdf = hist / sum(hist)  
    plt.plot(np.cumsum(pdf),'b',label=str(pareto_alpha1))
    
    hist, bin_edges = np.histogram(np.concatenate([np.zeros(1),np.min(dist2,axis=0),np.ones(1)*v]), density=True)
    pdf = hist / sum(hist)  
    plt.plot(np.cumsum(pdf),'g',label=str(pareto_alpha2))
    
    hist, bin_edges = np.histogram(np.concatenate([np.zeros(1),np.min(dist3,axis=0),np.ones(1)*v]), density=True)
    pdf = hist / sum(hist)  
    plt.plot(np.cumsum(pdf),'r',label=str(pareto_alpha3))

    if i==0:
        plt.legend()
plt.show()
```
